```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('athlete_events.csv')
```

```python
data_org = data.copy()
```

```python
data.head()
```

```python
Itemlist = []
for col in data.columns:
    Itemlist.append([col, data[col].dtype,
                    data[col].isnull().sum(),
                    round(((data[col].isnull().sum())/len(data[col]))*100,2),
                    data[col].nunique(),
                    list(data[col].sample(5).drop_duplicates().values)])
    
dfDesc = pd.DataFrame(columns=['dataFeatures', 'dataType', 'null', 'nullPct', 'unique', 'uniqueSample'], data=Itemlist)
dfDesc
```

```python
data.describe()
```

```python
name = data.Name
```

```python
name = name.tolist()
```

```python
import re
```

```python
def string_valid(string):
    regex = re.compile('[@_!#$%^&*""()<>?/\,|}{~:]') 
    if(regex.search(string) == None): 
        return True

    else: 
        return False
```

```python
Name = []
for s in name:
    sf = []
    for sub in s.split(' '):
        if(string_valid(sub)==1):
            sf.append(sub)
    Name.append(' '.join(sf))        
```

```python
data['Name'] = pd.Series(Name)
```

```python
team = data.Team
```

```python
Team = []
for s in team:
        Team.append(re.split(',/|_|-|!', s)[0])
```

```python
data['Team'] = pd.Series(Team)
```

```python
treat = ['Age','Height','Weight']
x = 1
plt.figure(figsize=(10,10))
for col in treat:
    plt.subplot(3,2,x)
    plt.title(col+' distribution')
    sns.distplot(data[col])
    
    plt.subplot(3,2,x+1)
    plt.title(col+' box plot')
    data[col].plot.box()
    print('{} skewness is : {}'.format(col, round(data[col].skew(),2)))
    x = x + 2
    
plt.tight_layout()
plt.show()
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Height'].fillna(data['Height'].median(), inplace=True)
data['Weight'].fillna(data['Weight'].median(), inplace=True)
```

```python
data.info()
```

```python
data['Medal'].fillna('None', inplace=True)
```

```python
# What is the highest number of sports played??

df = data[data['Year'].isin([1990, 1992, 1994, 2016])]
df.groupby('Year')['Sport'].value_counts().sort_values(ascending=False).groupby(['Year']).head(1).plot.bar(title='Highest no of sports played')
plt.show()
```

```python
# What are different sports played by female and which is the highest played sports

df = data[data['Sex'] == 'F']
print('Different types of sports played by female players are:\n')
print(df['Sport'].unique())

df['Sport'].value_counts().sort_values(ascending=False).head(10).plot.bar(title=' Top played sport by female players')
plt.show()
```

```python
# What are different sports played by male and which is the highest played sports

df = data[data['Sex'] == 'M']
print('Different types of sports played by male players are:\n')
print(df['Sport'].unique())

df['Sport'].value_counts().sort_values(ascending=False).head(10).plot.bar(title=' Top played sport by male players')
plt.show()
```

```python
# Which city has maximum number of players??

print(df['City'].value_counts().head(1))
print('\nLondon seems to have maximum number of players : {}'.format(df['City'].value_counts().head(1)[0]))
```

```python
# Show line chart which tells variation in number of players from 1990 to 2000 of India and Pakistan. (Draw separate chart for India and Pakistan) 

df = data[(data['Year']>= 1990) & (data['Year'] <= 2000) & (data['Team'].isin(['India', 'Pakistan']))]

# No of players from India by years
India = pd.DataFrame(df[df['Team'] == 'India']['Year'].value_counts()).reset_index()
India.rename(columns={'index' : 'Year', 'Year': 'Players'}, inplace=True)

# No of players from Pakistan by years
Pakistan = pd.DataFrame(df[df['Team'] == 'Pakistan']['Year'].value_counts()).reset_index()
Pakistan.rename(columns={'index' : 'Year', 'Year': 'Players'}, inplace=True)

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.title('No. of Indian players by years')
plt.plot(India['Year'], India['Players'])

plt.subplot(1,2,2)
plt.title('No. of Pakistan players by years')
plt.plot(Pakistan['Year'], Pakistan['Players'], 'g')

plt.show()
```

```python
# Draw pie chart to show percentage of male and female players across the world 

data['Sex'].value_counts(normalize=True).plot.pie(title='Percentage of male vs female participants', autopct = '%1.0f%%')
plt.show()
```

```python
# Which team has won maximum number of gold, silver and bronze. Visualize the same. 

df = data[data['Medal'] != 'None']
pd.crosstab(df['Team'],df['Medal']).sort_values(['Bronze','Silver','Gold'],ascending=(False,False,False)).head(1).plot.bar(title='Most medal winning team')
plt.show()
```

```python
# Who is tallest and shortest player ?

print('The tallest person ever to participate in olympics was {} of height : {}'.format(data[data['Height'] == data['Height'].max()]['Name'].unique()[0],data['Height'].max()))
print('The Shortest person ever to participate in olympics was {} of height : {}'.format(data[data['Height'] == data['Height'].min()]['Name'].unique()[0],data['Height'].max(),data['Height'].min()))
```

```python
# Draw a suitable chart which shows performance of different cities belonging to biejing, London, Sydney, Los Angeles. 

df = data[data['City'].isin(['Beijing', 'London', 'Sydney', 'Los Angeles'])]

df = pd.DataFrame(pd.crosstab(df['City'], df['Medal']))
df.drop('None', 1, inplace=True)
df.plot.bar(title='Performance of different cities')
plt.show()
```
