```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('athlete_events.csv')
```

```python
data.head(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Height'].fillna(data['Height'].median(), inplace=True)
data['Weight'].fillna(data['Weight'].median(), inplace=True)
data['Medal'].fillna('None', inplace=True)
```

```python
data.isnull().sum()
```

```python
df = data[data['Team'].isin(['United States', 'China', 'Netherlands', 'Spain', 'France', 'Italy'])]
```

```python
df['Team'].value_counts().plot.bar(title = 'Count of players from different countries')
plt.show()
```

```python
df.groupby('Sex')['Age'].mean().plot.bar(title='Average age by sex')
plt.show()
```

```python
df[df.Team == 'United States']['Sport'].unique()
```

```python
data[data['Medal'] != 'None']['Name']
```

```python
df= data.groupby('Season')['Sport'].unique()
print(df)
```

```python
df = data[data['Sport'].isin(['Football', 'Judo', 'Basketball'])]
print("Average Height of players is : {}".format(round(df['Height'].mean(),2)))
print("Average Weight of players is : {}".format(round(df['Weight'].mean(),2)))
```

```python
df = data[data['Games'].isin(['1992 Winter', '1994 Winter', '2000 Summer', '2002 Winter'])]
print(df.groupby('Games')['Sport','Event'].nunique())
```

```python
data[data['Medal'] != 'None']['Name'].tolist()
```

```python
df = data[data['Year'].isin([1990, 1992, 1994, 2016])]
df.groupby('Year')['Sport'].value_counts().sort_values(ascending=False).groupby(['Year']).head(1).plot.bar(title='Highest no of sports played')
plt.show()
```

```python
df = data[data['Sex'] == 'F']

df['Sport'].value_counts().sort_values(ascending=False).head(10).plot.bar(title=' Top played sport by female players')
plt.show()
```

```python
df = data[data['Sex'] == 'M']

df['Sport'].value_counts().sort_values(ascending=False).head(10).plot.bar(title=' Top played sport by male players')
plt.show()
```

```python
print(df['City'].value_counts().head(1))
```

```python
df = data[(data['Year']>= 1990) & (data['Year'] <= 2000) & (data['Team'].isin(['India', 'Pakistan']))]

# No of players from India by years
df_India = pd.DataFrame(df[df['Team'] == 'India']['Year'].value_counts()).reset_index()
df_India.rename(columns={'index' : 'Year', 'Year': 'Players'}, inplace=True)

# No of players from Pakistan by years
df_Pakistan = pd.DataFrame(df[df['Team'] == 'Pakistan']['Year'].value_counts()).reset_index()
df_Pakistan.rename(columns={'index' : 'Year', 'Year': 'Players'}, inplace=True)

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.title('No. of Indian players by years')
plt.plot(df_India['Year'], df_India['Players'])

plt.subplot(1,2,2)
plt.title('No. of Pakistan players by years')
plt.plot(df_Pakistan['Year'], df_Pakistan['Players'], 'g')

plt.show()
```

```python
data['Sex'].value_counts(normalize=True).plot.pie(title='Percentage of male vs female participants', autopct = '%1.0f%%')
plt.show()
```

```python
df = data[data['Medal'] != 'None']
pd.crosstab(df['Team'],df['Medal']).sort_values(['Bronze','Silver','Gold'],ascending=(False,False,False)).head(1).plot.bar(title='Most medal winning team')
plt.show()
```

```python
df = data[data['City'].isin(['Beijing', 'London', 'Sydney', 'Los Angeles'])]

df = pd.DataFrame(pd.crosstab(df['City'], df['Medal']))
df.drop('None', 1, inplace=True)
df.plot.bar(title='Performance of different cities')
plt.show()
```
